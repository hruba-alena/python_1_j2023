Stáhni si soubor platy_2021_02.csv

```python
import pandas
import numpy
import matplotlib.pyplot
```

Načti si tato data do tabulky 

```python
platy_2021_02 = pandas.read_csv("platy_2021_02.csv", index_col='cislo_zamestnance')  #names=['cislo_zamestnance', 'plat'],
platy_2021_02.head()
```

```python
platy_2021_02.plot(kind='bar', color='green', grid=True, figsize=(20, 5))
```

```python
serazene_platy=platy_2021_02.sort_values('plat')
serazene_platy.plot(kind='bar', color='green', grid=True, figsize=(20, 5))
```

vytvoř histogram - nastav vhodně hranice skupin histogramu (parametr `bins`), aby byl graf přehledný a snadno interpretovatelný.

```python
platy_2021_02.plot(kind='hist',grid=True, figsize=(10, 5), bins=[35000, 40000, 45000, 50000, 55000, 60000], xlabel='platy', edgecolor="red")

counts_1 = numpy.histogram(platy_2021_02['plat'], bins=[35000, 40000, 45000, 50000, 55000, 60000])
counts_1
```

```python
platy_2021_02.plot(kind='hist',grid=True, figsize=(10, 5), bins=5, xlabel='platy', edgecolor="red")
#serazeny_seznam_platy=sorted(platy_2021_02['plat'].tolist())
#serazeny_seznam_platy
#sloupec tabulky je vlastně seznam, nemusim prevadet na seznam:
counts_2 = numpy.histogram(platy_2021_02['plat'], bins=5)
counts_2
```
