Stáhni si soubor [temperature.csv](./temperature.csv), který obsahuje informace o průměrné teplotě v různých městech v **listopadu 2017**.

```python
teploty = pandas.read_csv("temperature.csv")
```


Vypiš si prvních několik řádků, ať si prohlédneš strukturu tabulky. 

```python
teploty.head(5)
```

Dotaz na měření, která byla provedena v Praze

```python
mereni_v_praze=teploty[teploty["City"] == "Prague"]
mereni_v_praze.head()
#len(mereni_v_praze)
```

 Dotaz na měření, ve kterých je teplota (sloupec `AvgTemperature`) vyšší než 80 stupňů F

```python
teploty_nad_80_F=teploty[teploty["AvgTemperature"] >80]
teploty_nad_80_F.head()
```

Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než -20 stupňů F

```python
extremni_teploty_F = teploty[(teploty["AvgTemperature"] < -20)| (teploty["AvgTemperature"] > 80)]
extremni_teploty_F.head(5)
len(extremni_teploty_F)
```

```python
teploty.sort_values(by=['AvgTemperature'], ascending=False)
```

prevod farenheita na celsia

```python
teploty["Celsia"]= ((5/9) * (teploty["AvgTemperature"] - 32))
#teploty.head()
teploty.to_csv('teploty_celsia.csv',index=False) 
teploty.sort_values(by=['Celsia'], ascending=True)
```

Dotaz na měření, ve kterých je teplota (sloupec Celsia) vyšší než 30 stupňů Celsia.

```python
teploty[(teploty["Celsia"] >30)]
teploty.head()
```

Dotaz na měření, ve kterých je teplota vyšší než 15 stupňů Celsia a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
teploty_Evropa_nad_15_C=teploty[(teploty["Celsia"] >15)&(teploty["Region"] =='Europe')]

#len(teploty_Evropa_nad_15_C)
```

```python
extremni_teploty_C= teploty[(teploty["Celsia"] < -10)| (teploty["Celsia"] > 30)]
extremni_teploty_C.head(5)
#len(extremni_teploty_C)
#max(extremni_teploty_C["Celsia"])
#min(extremni_teploty_C["Celsia"])
```
